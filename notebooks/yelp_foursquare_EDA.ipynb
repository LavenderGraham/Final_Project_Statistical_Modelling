```python
def get_bars_restaurants_lasertag_montreal_foursquare(locations, radius=1000, api_key=FOURSQUARE_KEY, categories='13003,13065,10022'):
    """
    Get venues from Foursquare for a list of locations within a specified radius and categories.
    """
    # Base URL for Foursquare Places API
    url = "https://api.foursquare.com/v3/places/search"
    
    # Create a list to hold all the venues
    all_venues = []
    
    # Loop over each location tuple in the list
    for latitude, longitude in locations:
        # Parameters for the API request
        params = {
            "ll": f"{latitude},{longitude}",
            "radius": radius,
            "categories": categories,
            "limit": 50  # You can adjust the limit as needed
        }
        
        # Headers including the authorization token
        headers = {
            "Accept": "application/json",
            "Authorization": api_key
        }
        
        # Sending GET request to the Foursquare API
        response = requests.get(url, headers=headers, params=params)
        
        if response.status_code == 200:
            # If the request was successful, parse the venues and add to the list
            venues = response.json().get('results', [])
            all_venues.extend(venues)
        else:
            # Handle the error
            print(f"Failed to retrieve data for location {latitude},{longitude}: Status code", response.status_code)
    
    # Return the aggregated list of venues
    return all_venues
```

```python
def get_businesses_yelp(locations, terms="bars,restaurants,lasertag", radius=1000, api_key=YELP_KEY):
    """
    Get businesses from Yelp for a list of locations within a specified radius and search term.
    """
    # Base URL for Yelp Business Search API
    url = "https://api.yelp.com/v3/businesses/search"
    
    # Create a list to hold all the businesses
    all_businesses = []
    
    # Loop over each location tuple in the list
    for latitude, longitude in locations:
        # Parameters for the API request
        params = {
            "latitude": latitude,
            "longitude": longitude,
            "radius": radius,
            "terms": terms,
            "limit": 50  # You can adjust the limit as needed
        }
        
        # Headers including the authorization token
        headers = {
            "accept": "application/json", "Authorization": f"Bearer {api_key}"}
        
        # Sending GET request to the Yelp API
        response = requests.get(url, headers=headers, params=params)
        print(f"Querying {latitude},{longitude}")
        
        if response.status_code == 200:
            # If the request was successful, parse the businesses and add to the list
            businesses = response.json().get('businesses', [])
            print(f"Found {len(businesses)} businesses")
            all_businesses.extend(businesses)
        else:
            # Handle the error
            print(f"Failed to retrieve data for location {latitude},{longitude}: Status code", response.status_code)
    
    # Return the aggregated list of businesses
    return all_businesses

```
