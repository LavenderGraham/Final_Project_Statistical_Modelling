```python
import numpy as np
import pandas as pd
import sqlite3
from sqlite3 import Error
```

```python
# Connect to an SQLite database (if the file does not exist, it will be created)
conn = sqlite3.connect('montreal_yelp_citybikes_database.sqlite')
```

```python
# Assuming 'conn' is your SQLite connection object and the query has already been executed

# Execute a query and fetch the results into a DataFrame
montreal_yelp_points_of_interest_df = pd.read_sql_query("SELECT * FROM points_of_interest", conn)
```

```python
montreal_yelp_points_of_interest_df_drop_null_price = montreal_yelp_points_of_interest_df.dropna(subset=['price'])
```

```python
import statsmodels.api as sm

X = montreal_yelp_points_of_interest_df['rating']
y = montreal_yelp_points_of_interest_df['Number of nearby bike stations']

```

```python
X = montreal_yelp_points_of_interest_df_drop_null_price['Number of nearby bike stations']
y = montreal_yelp_points_of_interest_df_drop_null_price['price']
```

```python
# Add a constant
X = sm.add_constant(X)
```

```python
# Create the model
model = sm.OLS(y, X).fit()
```

```python
# Get the summary of the regression
summary = model.summary()
print(summary)
```

# Stretch

How can you turn the regression model into a classification model?

```python
montreal_yelp_points_of_interest_df['Is Outstanding?'] = montreal_yelp_points_of_interest_df['rating'].apply(lambda x: 1 if x >= 4.5 else 0)


montreal_yelp_points_of_interest_df
```

```python
montreal_yelp_points_of_interest_df_drop_null_price["Exorbitant Price"] = montreal_yelp_points_of_interest_df_drop_null_price['price'].apply(lambda x: 1 if x >= 4.0 else 0)

montreal_yelp_points_of_interest_df_drop_null_price

```

```python
X = montreal_yelp_points_of_interest_df['Number of nearby bike stations']
y = montreal_yelp_points_of_interest_df['Is Outstanding?']
```

```python
X = montreal_yelp_points_of_interest_df_drop_null_price['Number of nearby bike stations']
y = montreal_yelp_points_of_interest_df_drop_null_price['Exorbitant Price']

```

```python
# Add a constant
X = sm.add_constant(X)

```

```python
# Initialize and fit the logistic regression model
logit_model = sm.Logit(y, X).fit()

```

```python
# Display the model summary
print(logit_model.summary())

```

```python
#This logistic regression has found that the number of bike stations near an establishment has a statistically significant negative association with whether or not it is outstanding in rating (4.5 stars or more)
#but the correlation is very weak.
```

```python
# Display the model summary
print(logit_model.summary())
```

```python
#This logistic regression has found that the number of bike stations near an establishment has a statistically significant association association with whether or not it is exhorbiantly priced ($$$$ on Yelp)
#but the correlation is very weak.
```
